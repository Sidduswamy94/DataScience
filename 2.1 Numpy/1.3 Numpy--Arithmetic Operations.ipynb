# <p style='text-align: center;'> Numpy--Arithmetic Operations (OR) Matrix Operations </p>

- On the objects of ndarray, we can apply all types of Arithmetic Operators.


- To perform Arithmetic Operations on the objects of ndarray in numpy programming, we use the following functions.

            a) add()
            b) subtract()
            c) multiply()
            d) dot() or matmul()
            e) divide()
            f) floor_divide()
            g) mod()
            h) power()
            
            
- All the arithmetic Functions can also be performed w.r.t Arithmetic Operators.


- All these Arithmetic Operations are called Matrix Operations.


-----------------------
### a) add() :
-----------------------
- <b> Syntax: -   varname=numpy.add(ndarrayobj1, ndarrayobj2)
    
    
- This function is used for adding an elements of ndarrayobj1, ndarrayobj2 and result can be displayed.
    
    
- We can use "+" operator for addition purpose. It will also work same as add().


```python
# Python Programming for domenstrating add() function in numpy.
 
# importing the numpy module as alias name.
import numpy as np

# creating two list for addition.
l1=[[10,20],[30,40]]
l2=[[1,2],[3,4]]

# create the two ndarray object.
# pass the lists in ndarray object.
a=np.array(l1)        # passing "l1" in "a" ndarray object.
b=np.array(l2)        # passing "l2" in "b" ndarray object.

# print the a and b objects.
print("Containts of a object:")
print(a)              # printing object "a"
print("-"* 50)
print("-"* 50)
print("Containts of b object:")
print(b)              # printing object "b"
print("-"* 50)
print("-"* 50)

# calculating addition of two ndarray objects.
c=np.add(a,b)         # Here "c" is a variable

# Printing addition of two objects. 
print("Addition of a and b objects:")
print(c)              # it will returns the addition of two ndarray objects.
```

```python
''' Python Programming for domenstrating addition of two ndarray objects in 
    numpy by using "+" operator.'''
 
# importing the numpy module as alias name.
import numpy as np

# creating two list for addition.
l1=[[100,200],[300,400]]
l2=[[10,20],[30,4]]

# create the two ndarray object.
# pass the lists in ndarray object.
a=np.array(l1)        # passing "l1" in "a" ndarray object.
b=np.array(l2)        # passing "l2" in "b" ndarray object.

# print the a and b objects.
print("Containts of a object:")
print(a)              # printing object "a"
print("-"* 50)
print("-"* 50)
print("Containts of b object:")
print(b)              # printing object "b"
print("-"* 50)
print("-"* 50)

# calculating addition of two ndarray objects by "+" operator.
# we used operator "+" instead of add().
c=a+b         # Here "c" is a variable

# Printing addition of two objects. 
print("Addition of a and b objects:")
print(c)              # it will returns the addition of two ndarray objects.
```

### d) dot() or matmul() :
------------------------------
- To perform Matrix Multiplication, we use dot(), matmul().


- <b> Syntax: -    varname=numpy.dot(ndarrayobj1, ndarrayobj2)
    
    
    
- <b> Syntax: -    varname=numpy.matmul(ndarrayobj1, ndarrayobj2)
    
    
- These functions is used for performing actual matrix multiplication of ndarrayobj1, ndarrayobj2 and result can be 
  displayed.
    
    
    
- This function works as shown in below image.   



```python
# Python Programming for domenstrating dot() function in numpy.
 
# importing the numpy module as alias name.
import numpy as np

# creating two list.
l1=[[10,20],[30,40]]
l2=[[1,2],[3,4]]

# create the two ndarray object.
# pass the lists in ndarray object.
a=np.array(l1)        # passing "l1" in "a" ndarray object.
b=np.array(l2)        # passing "l2" in "b" ndarray object.

# print the a and b objects.
print("Containts of a object:")
print(a)              # printing object "a"
# Shape of a array is:  (2, 2)
print("-"* 50)
print("-"* 50)
print("Containts of b object:")
print(b)              # printing object "b"
# Shape of b array is:  (2, 2)
print("-"* 50)
print("-"* 50)

# calculating matrix multiplication of two ndarray objects.
c=np.dot(a,b)         # Here "c" is a variable

# Printing matrix multiplication of two objects. 
print("Matrix Multiplication of a and b objects:")
print(c)      # it will returns the matrix multiplication of two ndarray objects.
# Shape of c array is:  (2, 2)
```

```python
# Python Programming for domenstrating matmul() function in numpy.
 
# importing the numpy module as alias name.
import numpy as np

# creating two list.
l1=[[5,10],[3,4]]
l2=[1,2]

# create the two ndarray object.
# pass the lists in ndarray object.
a=np.array(l1)        # passing "l1" in "a" ndarray object.
b=np.array(l2)        # passing "l2" in "b" ndarray object.

# print the a and b objects.
print("Containts of a object:")
print(a)              # printing object "a"
# Shape of a array is:  (2, 2)
print("-"* 50)
print("-"* 50)
print("Containts of b object:")
print(b)              # printing object "b"
# Shape of b array is:  (1, 2)
print("-"* 50)
print("-"* 50)

# calculating matrix multiplication of two ndarray objects.
c=np.matmul(a,b)         # Here "c" is a variable

# Printing matrix multiplication of two objects. 
print("Matrix Multiplication of a and b objects:")
print(c)      # it will returns the matrix multiplication of two ndarray objects.
# Shape of c array is:  (1, 2)
```
