## Histogram / Distribution Plot :
- It is used basically for univariant set of observations and visualizes it through a histogram.


- A histogram shows continuous data in ordered rectangular columns. Usually, there are no gaps between the columns.


- The histogram displays a frequency distribution (shape) of a data set. At first glance, histograms look alike to bar graphs.   However, there is a key difference between them. Bar Chart represents categorical data and histogram represent continuous 
  data.
  
  
- Histogram uses when you want to see whether the outputs of two or more processes are different.


- Histogram uses to check whether data is skewed or normal and to check how the data is distributed.


- Histogram is a continous frequency distribution and Univariate Analysis plot.

-------------------
<b> Examples 
-------------------

```python
'''WAPP to plot the histogram represents count / frequency Vs total_bill 
   from "df" dataframe.'''

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

'''Plot a simple histogram represents count / frequency Vs total_bill 
   from "df" dataframe, Set bins=10, Set kde=True.'''
sns.displot(df["total_bill"], bins=10, kde=True)

# Display all figures.
plt.show()
```

<b> In the above example :
- KDE stands for Kernel Density Estimation and that is another kind of the plot in seaborn.


- bins is used to set the number of bins you want in your plot and it actually depends on your dataset.


<b> Now looking at this we can say that most of the total_bill given lies between 10 and 20.

## Box Plot :
- A box and whisker plot is a graphical representation of the collected data on a number line with the use of the least value, greatest value and the quartiles.


- The Box and Wisker plot is mainly used to determine the outliers from the given data set.


- In the Box Plot, the x-axis indicates the data to be plotted while the y-axis denotes the frequency distribution.


- Box Plot is a continous frequency distribution and Univariate Analysis plot.


- Box Plot used on continous data.


- Box Plot is also know as Five Number Summary.


- Quartiles divide the the data into four equal parts, the first quartile is the 25th percentile, second is the 50th percentile and the third quartile is the 75th percentile of the data.

-------------------
<b> Examples :
-------------------

```python
# WAPP to plot the Box Plot on tip variable from "df" dataframe.

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

# Plot the Box Plot on tip variable from "df" dataframe.
# Set boxplot on y-axis (Vertically)
sns.boxplot(y=df["tip"])              

# Display all figures.
plt.show()
```

## Scatter Plot : -
- A scatter plot is a diagram where each value in the data set is represented by a dot.


- The scatter plot is an X-Y diagram that shows a relationship between two variables.


- It is used to plot data points on a vertical and a horizontal axis. The purpose is to show how much one variable affects 
  another.
  
  
- Usually, when there is a relationship between 2 variables, the first one is called independent. The second variable is called 
  dependent because its values depend on the first variable.
  
  
- Scatter plots also help you predict the behavior of one variable (dependent) based on the measure of the other variable 
  (independent).
  
  
- Scatter plot uses when having paired numerical data.


- Scatter plot uses when working with root cause analysis tools to identify the potential for problems.


- Scatter plot uses when you just want to visualize the correlation between 2 large datasets without regard to time.


- Scatter plot is considering 2 continous variables and it is a Bivariate Analysis plot.


- Scatter plot uses to check the i) Direction  ii) Strength   iii) linearity of the variables.

-------------------------
<b> Examples :
-------------------------

```python
'''WAPP to plot the Scatter plot on total_bill Vs tip variables 
   from "df" dataframe.'''

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

# Plot the Scatter plot on total_bill Vs tip variables from "df" dataframe.
# Set "total_bill" variable on x-axis.
# Set "tip" variable on y-axis.
sns.scatterplot(x=df["total_bill"], y=df["tip"])

# Display all figures.
plt.show()
```

# Violin Plot : -
- Violin Plot helps us to see both the distribution of data in terms of Kernal Density estimation and the box plot.


- A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data.


- It is used to visualize the distribution of numerical data.


- Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each 
  variable.


- It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those 
  distributions can be compared.

--------------------
<b> Examples :
--------------------

```python
# WAPP to plot the violinplot on total_bill Vs day variables from "df" dataframe.

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

# plot the violinplot on total_bill Vs day variables from "df" dataframe.
# Set "total_bill" variable on x-axis.
# Set "day" variable on y-axis.
# Set data=df.
sns.violinplot(x="total_bill", y="day", data=df)

# Display all figures.
plt.show()
```

<b> In the above example :
- The white dot represents the median.


- The thick gray bar in the center represents the interquartile range.
    
    
- the thin gray line represents the rest of the distribution, except for points that are determined to be “outliers” using a 
  method that is a function of the interquartile range.
    
    
- On each side of the gray line is a kernel density estimation to show the distribution shape of the data. Wider sections of 
  the violin plot represent a higher probability that members of the population will take on the given value; the skinnier 
  sections represent a lower probability.
    
    
- This violin plot shows the relationship of total_bill Vs day. The box plot elements show the median total_bill for 
  Fri-day is lower than for other day types. The shape of the distribution (extremely skinny on each end and wide 
  in the middle) indicates the total_bill of sat-day is highly concentrated around the median.

## Pair Plot : -
- A Pair Plot is also known as scotter plot, in which one variable in the same data row is matched with another variables 
  value, like this pair plots are just elaborations on this, showing all variables paired with all the other variables.
  
  
- Pairplot visualizes given data to find the relationship between them where the variables can be continuous or categorical.


- Pairplot is a module of seaborn library which provides a high-level interface for drawing attractive and informative 
  statistical graphics.

--------------------
<b> Examples :
--------------------

```python
# WAPP to plot the pairplot on continous variables from "df" dataframe.

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

# Store continous data variable seperately in "continous" object.
continous = ["total_bill", "tip"]

# plot the pairplot on continous variables from "df" dataframe.
# Set vars = continous
sns.pairplot(df, vars=continous)

# Display all figures.
plt.show()
```

  to display the plot to the users. As a result, the pair plot has been generated successfully.

## Heat Map : -
- A heatmap uses colored cells to represent relation between variables.


- To represent more common values or higher activities brighter colors basically reddish colors are used and to represent less 
  common or activity values, darker colors are preferred.


<b> Heat Map (for correlation)
- A correlation heatmap uses colored cells to show a 2D correlation matrix (Table) between two numeric dimensions.
    
    
- The values of the first dimension appear as the rows of the table while of the second dimension as a column.
    
    
- The color of the cell is proportional to the number of measurements that match the dimensional value. This makes correlation 
  heatmaps ideal for data analysis since it makes patterns easily readable and highlights the differences and variation in the 
  same data.
    
    
- It is very important in Feature Selection.
    
    
                                   Cov(X,Y)
            Correlation, ρ(X,Y) = ---------
                                    σX σy
    

--------------------
<b> Examples :
--------------------

```python
# WAPP to plot the heatmap on total_bill and tip variables from "df" dataframe.

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

'''calculate the correlation between total_bill and tip variables and store it 
   in "c_m" object.'''
c_m = df[["total_bill","tip"]].corr()

# plotting correlation heatmap.
# Set annot=True.
sns.heatmap(c_m, annot=True)

# displaying heatmap
plt.show()
```

<b> In the above example :
- This heatmap represent higher activities as correlation = 1, brighter colors basically reddish colors are used and to 
  represent less activity values as correlation = 0.68, darker colors are used.
    
    
- annot : If True, write the data value in each cell.
    
    
- On the color of heatmap we can easily aidentify the relation between variables.

- <b> Syntax :

        sns.catplot(data= None, x=None, y=None, hue=None, color=None, kind= 'fill with your plot type')
    
    
- Where,
    
    - x, y, hue :  names of variables in data / Inputs for plotting long-form data.
    - data      :  DataFrame name (In which dataframe do you want to plot).
    - kind      :  fill with your plot type (str), optional.
    - color     :  matplotlib color, optional.

## Box Plot :
- A box and whisker plot is a graphical representation of the collected data on a number line with the use of the least value, greatest value and the quartiles.


- The Box and Wisker plot is mainly used to determine the outliers from the given data set.


- In the Box Plot, the x-axis indicates the data to be plotted while the y-axis denotes the frequency distribution.


- Box Plot is a continous frequency distribution and Univariate Analysis plot.


- Box Plot used on continous data.


- Box Plot is also know as Five Number Summary.


- Quartiles divide the the data into four equal parts, the first quartile is the 25th percentile, second is the 50th percentile 
  and the third quartile is the 75th percentile of the data.

```python
# WAPP to plot the boxplot on day and total_bill variables from "df" dataframe.

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

# plot the boxplot on day and total_bill variables from "df" dataframe.
# Set x="day".
# Set y="total_bill"
# Set data=df
sns.boxplot(x="day", y="total_bill", data=df)

# Show the plot
plt.show()
```

- <b> From the above example, we can easily observed that the outliers are there in each day Vs total_bill dataframe.

```python
# WAPP to plot the catplot on day and total_bill variables from "df" dataframe.

# import the pandas as pd.
import pandas as pd

# import the matplotlib.pyplot as plt.
import matplotlib.pyplot as plt

# import the seaborn as sns.
import seaborn as sns

# Plot the catplot with x="day", y="total_bill", data=df, kind="box" and hue="smoker".
sns.catplot(x="day", y="total_bill", data=df, kind="box", hue="smoker")

# Show the plot
plt.show()
```

- <b> From the above example we can read the data easily, in which data outliers are there or not.
