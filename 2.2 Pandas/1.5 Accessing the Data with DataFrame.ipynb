### Working with Duplicate values

<b> duplicated() :
- The duplicated() method returns a Series with True and False values that describe which rows in the DataFrame are 
  duplicated and not.
    
    
    
- <b> Syntax : - dataframeobj.duplicated(subset=None, keep='first')
    
    

- Here subset : (default: None). It is used to specify the particular columns in which duplicate values are to be searched.       subset is optional.
        
        
- keep:‘first’ or ‘last’ or False (default: ‘first’). It is used to specify which instance of the repeated rows is to be 
  identified as a unique row. keep is also optional.



- It will returns a Boolean series where the value True indicates that the row at the corresponding index is a duplicate and 
  False indicates that the row is unique.
    
    
    
- Use the subset parameter to specify if any columns should not be considered when looking for duplicates.



```python
# WAPP to demonstrate with duplicated().

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling duplicated() function. 
df.duplicated()
```

- In the above example, we got duplicate values at row_index=2.


- In the above example we have not used subset and keep parameters, it takes default params.

```python
# WAPP to demonstrate duplicated() with single column.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling duplicated() function. 
# pass subset="Age" and keep="last"
df.duplicated(subset="Age", keep="last")
```

- Here we have checked dulpicates with single column "Age" and repeated rows considered 1st as dulpicate and 2nd as  
  unique.



- In the above example, we got duplicate values at index=1

<b> drop.duplicates() :
- The drop_duplicates() method removes duplicate rows.



- <b> Syntax : - DataFrameobj.drop_duplicates(subset=None, keep=’first’, inplace=False)
    
    

- subset (Optional) : Subset takes a column or list of column label. It’s default value is none. After passing columns, 
  it will consider them only for duplicates. 
   
    
    
- Use the "subset" parameter if only some specified columns should be considered when looking for duplicates.
    
    
    
- keep (Optional) : keep is to control how to consider duplicate value. It has only three distinct value and default is first’. 
    
    
    
- a)bIf ‘first‘, it considers first value as unique and rest of the same values as duplicate.
    
    
    
    
- b) If ‘last‘, it considers last value as unique and rest of the same values as duplicate.
    
    
    
- c) If False, it consider all of the same values as duplicates.
    
    
    
- inplace (Optional) : default False. If True: the removing is done on the current DataFrame. If False: returns a copy where 
  the removing is done.

```python
# WAPP to domenstrate drop_duplicates() method.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling drop_duplicated() method. 
'''Here subset, keep, inplace are optional, we are neglecting these values
   and considering as by default (subset=None, keep=’first’, inplace=False).''' 
df.drop_duplicates()  # By default, it removes duplicate rows based on all columns.
```

- In the above example, we can see the row_index=1 and 2 both having same data, here row_index=1 is considered as unique value
  and row_index=2 is considered as duplicate value, so row_index=2 has been deleted.
  
  
  
- By default, it removes duplicate rows based on all columns.

```python
# WAPP to remove duplicates on specific column(s) by using drop_duplicates() method.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling drop_duplicated() method. 
'''Here subset="Age", here we are neglecting keep and inplace values
   and considering as by default (keep=’first’, inplace=False).''' 
df.drop_duplicates(subset="Age")  # By default, it removes duplicate rows based on "Age" column.
```

- In the above example, we can see the row_index=1 and 2 are containts Age=62 both having same age and row_index=7 and 9 are 
  containts Age=29 both having same age, here row_index=1 and 7 are considered as unique values and row_index=2 and 9 are 
  considered as duplicate value, so row_index=2 and 9 has been deleted.

### Working with drop rows or columns from dataFrame :

drop() :
- The drop() method removes the specified row or column.


- By specifying the column axis (axis='columns' or axis=1), the drop() method removes the specified column.


- By specifying the row axis (axis='index' or axis=0), the drop() method removes the specified row.


- <b> Syntax : - dataframeobj.drop(labels, axis, index, columns, inplace)
    
    
- Here labels (optional) : The labels or indexes to drop. If more than one, specify them in a list.
    
    
- axis (optional) : Which axis to check, for row axis (axis='index' or axis=0) ans for column axis (axis='columns' or axis=1),
  default 0.
    
    
- index (Optional) : Specifies the name of the rows to drop. Can be used instead of the labels parameter.
    
    
- columns (optional) : Specifies the name of the columns to drop. Can be used instead of the labels parameter.
    
    
- inplace (optional) : default False. If True: the removing is done on the current DataFrame. If False: returns a copy where 
  the removing is done.

```python
# WAPP to drop the single column by using drop() method.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling drop() method.
# pass the column name and axis='columns' or axis=1
df.drop("DB", axis=1)   # It will drop the "DB" column.
```

```python
# WAPP to drop the multiple columns by using drop() method.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling drop() method.
# specify the column names in list pass the list and axis='columns' or axis=1.
df.drop(["TB","TP"], axis=1)   # It will drop the "TB" and "TP" columns.
```

- In the above examples, if you want to store the values after dropping the columns in the original or new dataframe, we need

<b> sort_values() :
- The sort_values() method sorts the DataFrame by the specified label.
    
    
- <b> Syntax : -  dataframeobj.sort_values(by, axis, ascending, inplace)
    
    
- Here by (Required) : Specify labels to sort by. Either index level(s) or column label(s). Or, if the axis is 1 or 'columns' this values specify column level(s) or index label(s)
    
    
- axis (optional) : Specifies the axis to sort by, for row axis (axis='index' or axis=0) ans for column axis (axis='columns' or axis=1), default 0.
    
    
- ascending (optional) : default True. Specifies whether to sort ascending (0 -> 9) or descending (9 -> 0)
    
    
- inplace (optional) : default False. If True: the removing is done on the current DataFrame. If False: returns a copy where 
  the removing is done.

```python
# WAPP to demonstrate the sort_values() method in ascending order.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling sort_values() method.
# pass the by="column name" and ascending=True.
df.sort_values(by="DB", ascending=True)  # It will sort in ascending order based on DB column.
```

<b> crosstab() :
- This method is used to compute a simple cross-tabulation of two (or more) factors.


- By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed.

```python
# WAPP to demonstrate crosstab() method with "LiverPatient" and "Gender" columns.

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling crosstab() method with Gender and LiverPatient.
''' find how many Female's and Male's with LiverPatient "Yes" and "No"  
    count seperatally by using crosstab() method.'''
pd.crosstab(df.LiverPatient, df.Gender)
```

```python
''' WAPP to demonstrate crosstab() method with "LiverPatient" and "Gender" 
    columns and find the count of Gender LiverPatient totally.'''

# importing pandas as alias_name pd
import pandas as pd

# read csv file into pandas DataFrame.
df=pd.read_csv("liver.csv")

# calling crosstab() method with Gender and LiverPatient.
''' find how many Female's and Male's with LiverPatient "Yes" and "No"  
    count seperatally by using crosstab() method.'''
# pass margins=True for total count of LiverPatient with Gender.
pd.crosstab(df.LiverPatient, df.Gender, margins=True)
```
