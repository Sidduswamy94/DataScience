<b> Here we are using "homeprices.csv" dataset for Train_Test Split.
    

```python
# import the required libraries (modules).
import numpy as np
import pandas as pd


# Load the Dataset.
df = pd.read_csv("homeprices.csv")

# print the DataFrame.
df
```

In the above we have imported the required libraries (modules), then loaded the "homeprices.csv" dataset and printed that on console.


```python
# Display dimentionality of the DataFrame.
df.shape
```

Here 12 rows (records) and 3 columns (variables) are their.


We can see that there is no null values in this dataset.


In the above data, there are 2 continuous variables and 1 categorical variable is their. So we need to encode (convert) the categorical into Numeric first. Here "town" variable category is nominal, so we are going with nominal encoding on this feature.


```python
# Apply the nominal encoding by using pd.get_dummies() function.
# And store it in "df_dum" object.
df_dum = pd.get_dummies(df,drop_first=True)

# Print the result.
df_dum
```

In the above, we have applied the nominal encoding by using pd.get_dummies() function and dropped the first dummies i.e. to get k-1 dummies out of k categorical levels by removing the first level. And printed that on console.


From the above example the town variable is converted to dummy variables, the three vectors obtained as Bangalore, Chennai,
Hyderabad. In the 1st row a 1 in the "Chennai" column would mean a 0 in the "Bangalore" and "Hyderabad" column, we can say X(Chennai) = 1 - X(Bangalore) - X(Hyderabad). This results in three multi-collinear dummy variables, so the dummy variable trap may occur in regression analysis.

```python
# Create x variable as independent variables.
x = df_dum.drop("price", axis="columns")

# Print the x variable.
x
```

In the above, we have created x variable with 3 independent (input) variables and dropped one variable i.e dependent variable and printed x variables on console.

```python
# Create x variable as dependent variables.
y = df_dum["price"]

# Print the y variable.
y
```

In the above, we have created y variable with 1 dependent (output) variable and printed y variable on console.

```python
# import train_test_split from scikit-learn.
from sklearn.model_selection import train_test_split

# Apply the train_test_split() function.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=9)
```

In the first line of the above code, we have imported the train_test_split function from the sklearn library.


In the second line, we have used four variables, which are :

   - x_train: It is used to represent features for the training data
   - x_test: It is used to represent features for testing data
   - y_train: It is used to represent dependent variables for training data
   - y_test: It is used to represent independent variable for testing data
   
   
In the train_test_split() function, we have passed four parameters. Which first two are for arrays of data, and test_size is for specifying the size of the test set which tells the dividing ratio of training and testing sets. The last parameter, random_state, is used to set a seed for a random generator so that you always get the same result.
