## Reading/Loading, Displaying and Writing/Save Images

### imread():
 
- The imread() function loads image from the specified file and returns it.


<b> The syntax is:  cv2.imread(filename[,flag]) 

Where, Parameters are:

- **filename:** Name of the file to be loaded


- **flag:** The flag specifies the color type of a loaded image
    
    
- **CV_LOAD_IMAGE_ANYDEPTH** - If we set it as flag, it will return 16-bits/32-bits image when the input has the corresponding depth, otherwise convert it to 8-BIT.
    
    
- **CV_LOAD_IMAGE_COLOR** - If we set it as flag, it always return the converted image to the color one.
    
    
- **C V_LOAD_IMAGE_GRAYSCALE** - If we set it as flag, it always convert image into the grayscale.
    

The **imread()** function returns a **matrix**, if the image cannot be read because of unsupported file format, missing file, unsupported or invalid format. Currently, the following file formats are supported.

- **Window bitmaps** - *.bmp, *.dib
- **JPEG files** - *.jpeg, *.jpg, *.jpe
- **Portable Network Graphics** - *.png
- **Portable image format** - *.pbm, *.pgm, *.ppm
- **TIFF files** - *.tiff, *.tif    
    
    
**Note:** The color images, the decoded images will have the channels stored in the BGR order.

<b> The syntax is the following: cv2.imwrite(filename, img[,params])

### Making Borders for Images
- OpenCV provides the **cv2.copyMakeBorder()** function to create a border around the image, something like a photo frame. 


- <b> The syntax is: cv2.copyMakeBorder(src,top,bottom,left,right,border type)
    
    
Where, Parameters are: 
    
- **src** - It denotes input image.

    
- **top, bottom, left, right** - It defines the border width in the number of pixels in the corresponding direction.

    
- **borderType** - It defines what kind of border to be added. The border can be the following types.

    
- **value** - Color of border if border type is cv.BORDER_CONSTANT    
    
    
<b> Consider the following example:    

```python
import cv2 as cv  
import numpy as np  
from matplotlib import pyplot as plt  
BLUE = [250,0,0]  
c_img = cv2.imread('Dataset/image_examples/Modi.jpg',1) 
replicate = cv.copyMakeBorder(c_img,10,10,10,10,cv.BORDER_REPLICATE)  
reflect = cv.copyMakeBorder(c_img,10,10,10,10,cv.BORDER_REFLECT)  
reflect101 = cv.copyMakeBorder(c_img,10,10,10,10,cv.BORDER_REFLECT_101)  
wrap = cv.copyMakeBorder(c_img,10,10,10,10,cv.BORDER_WRAP)  
constant= cv.copyMakeBorder(c_img,10,10,10,10,cv.BORDER_CONSTANT,value=BLUE)  
plt.subplot(231),plt.imshow(c_img,'gray'),plt.title('ORIGINAL')  
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')  
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')  
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')  
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')  
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')  
plt.show()  
```

### Image Rotation
- The image can be rotated in various angles (90,180,270 and 360). OpenCV calculates the affine matrix that performs affine transformation, which means it does not preserve the angle between the lines or distances between the points, although it preserves the ratio of distances between points lying on the lines.


- <b> The syntax is: cv2.getRotationMatrix2D(center, angle, scale rotated = cv2.warpAfifne(img,M,(w,h))
    
    
Where, Parameters are:
    
- **center:** It represents the center of the image.
    
    
- **angle:** It represents the angle by which a particular image to be rotated in the anti-clockwise direction.
    
    
- **rotated:** ndarray that holds the rotated image data.
    
    
- **scale:** The value 1.0 is denoted that the shape is preserved. Scale the image according to the provided value. 
    

<b> Consider the following example:    

```python
import cv2  
import matplotlib.pyplot as plt 

# read image as color image  
img = cv2.imread('Dataset/image_examples/cat_or_dog_1.jpg') 

# resize the image 
resized_image = cv2.resize(img, (200, 200))

# get image height, width  
(h, w) = resized_image.shape[:2]  

# calculate the center of the image  
center = (w / 2, h / 2)  

angle90 = 90                         # angle
scale = 1.0                          # scale

# Perform the counterclockwise rotation holding at the center  
# 90 degrees  
M = cv2.getRotationMatrix2D(center, angle90, scale)  
rotated90 = cv2.warpAffine(resized_image, M, (h, w))  

cv2.imshow('Original Image', resized_image)  
cv2.waitKey(0)                       # waits until a key is pressed  
cv2.destroyAllWindows()              # destroys the window showing image  
  
cv2.imshow('Image rotated by 90 degrees', rotated90)  
cv2.waitKey(0)                       # waits until a key is pressed  
cv2.destroyAllWindows()              # destroys the window showing imag 
```

## <p style='text-align: center;'> Face Detection using HAAR Cascade Classifier </p>

### Face Detection using HAAR Cascade Classifier:
- OpenCV supports haar cascade based object detection. Haar cascades are machine learning based classifiers that calculate different features like edges, lines, etc in the image.


- In the below, we are going to display a list of rectangles of all the detected objects (faces in our first case) by using **detectMultiScale()** method.   

```python
#import required libraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

# load the 'haarcascade_frontalface_default.xml' file to classifiers as 'face_classifier'
face_classifier = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_frontalface_default.xml')

# read the image, resize the image and convert to grayscale format
image = cv2.imread('Dataset/image_examples/Modi.jpg')
image = cv2.resize(image, (500, 500))
gray = cv2.cvtColor(image, cv.COLOR_BGR2GRAY)

# calculate coordinates 
# Tuning Cascade Classifier - detectMultiScale('Input Image', "Scale Factor", "Minimum Neighbors")
faces = face_classifier.detectMultiScale(gray, 1.05, 5)

# print the result
print(faces)
```

In the above, first we have imported the required libraries, then loaded the 'haarcascade_frontalface_default.xml' file to classifiers as 'face_classifier' by using '**cv2.CascadeClassifier()**'. '**cv2.CascadeClassifier()**' is a machine learning based approach where a cascade function is trained from a lot of positive and negative images. It is then used to detect objects in other images. After that loaded the 'Modi.jpg' image and stored as 'image', then resized the image as (200, 200) by using '**cv2.resize**' and same result stored as 'image' object. Then converted the color image to grayscale format by using '**cv2.cvtColor**'. 


After that calculated coordinates (Tuning Cascade Classifier) by using '**detectMultiScale**' class. The **detectMultiScale()** method returns a list of rectangles of all the detected objects (faces in our first case). Each element in the list represents a unique face. This list contains tuples, (x, y, w, h), where the x, y values represent the top-left coordinates of the rectangle, while the w, h values represent the width and height of the rectangle, respectively. In **detectMultiScale()** class we have passed some parameters they are:

- **image:** Matrix of the type CV_8U containing an image where objects are detected.


- **scaleFactor:** Parameter specifying how much the image size is reduced at each image scale.


- **minNeighbors:** Parameter specifying how many neighbors each candidate rectangle should have to retain it.


As we can see that, it returns a list of rectangles of all the detected objects (faces in our first case). Each element in the list represents a unique face. This list contains tuples, (x, y, w, h), where the x, y values represent the top-left coordinates of the rectangle, while the w, h values represent the width and height of the rectangle, respectively.


<b> Now in the below, we are going to detect the face from the given image.

```python
#import required libraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter("ignore")

# load the 'haarcascade_frontalface_default.xml' file to classifiers as 'face_classifier'
face_classifier = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_frontalface_default.xml')

# read the image, resize the image and convert to grayscale format
image = cv2.imread('Dataset/image_examples/Modi.jpg')
image = cv2.resize(image, (500, 500))
gray = cv2.cvtColor(image, cv.COLOR_BGR2GRAY)

# calculate coordinates 
# Tuning Cascade Classifier - detectMultiScale('Input Image', "Scale Factor", "Minimum Neighbors")
faces = face_classifier.detectMultiScale(gray, 1.05, 5)

# if no faces detected, then retuns the empty tuple.
if faces is ():
    print('No faces found')
    
    
# if faces detected, we iterate through our faces array and draw a rectangle over each face in faces.
# draw bounding boxes around detected features
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,100), 1)
    

#plot the image
plt.imshow(image)
```

<b> As we can see that, in the above we have detected the face with rectangular box from the given image.
    


### Face & Eye Detection using HAAR Cascade Classifier in given Image:
<b> Now in the below, we are going to detect the face and eye from the given image.

```python
# import required libraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter("ignore")

# load the 'haarcascade_frontalface_default.xml' file to classifiers as 'face_classifier'
face_classifier = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_frontalface_default.xml')

# load the 'haarcascade_eye.xml' file to classifiers as 'eye_classifier'
eye_classifier = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_eye.xml')

# read the image, resize the image and convert to grayscale format
image = cv2.imread('Dataset/image_examples/Modi.jpg')
resized_image = cv2.resize(image, (500, 500))
gray = cv2.cvtColor(resized_image, cv.COLOR_BGR2GRAY)

# calculate coordinates 
# Tuning Cascade Classifier - detectMultiScale('Input Image', "Scale Factor", "Minimum Neighbors")
faces = face_classifier.detectMultiScale(gray, 1.3, 5)

# if no faces detected, then retuns the empty tuple.
if faces is ():
    print('No faces found')
    
    
# if faces detected, we iterate through our faces array and draw a rectangle over each face in faces.
# draw bounding boxes around detected features and draw a rectangle over each eye in faces. 
for (x,y,w,h) in faces:
    cv2.rectangle(resized_image, (x,y), (x+w, y+h), (255,0,0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = resized_image[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(roi_gray)
    
    # draw bounding boxes around detected features
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
    

# plot the image
plt.imshow(resized_image)
```

<b> In the above, we have detected the face and eyes with rectangular box from the given image.

## <p style='text-align: center;'> Working with VideoCapture/Videos </p>

### OpenCV VideoCapture
- OpenCV provides the **VideoCature()** function which is used to work with the Camera.


<b> Capture Video from Camera:
    
- OpenCV allows a straightforward interface to capture live stream with the camera (webcam). It converts video into grayscale and display it.

    
- We need to create a VideoCapture object to capture a video. It accepts either the device index or the name of a video file. A number which is specifying to the camera is called device index. We can select the camera by passing the O or 1 as an argument. After that we can capture the video frame-by-frame.

```python
import cv2  
import numpy as np  

# Get a reference to webcam 
cap = cv2.VideoCapture(0)  
  
while(True):  
    # Capture image frame-by-frame  
    ret, frame = cap.read()  
  
    # Our operations on the frame come here  
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  
  
    # Display the resulting frame  
    cv2.imshow('frame',gray)  
    if cv2.waitKey(1) == ord('q'):  
        break  
  
# When everything done, release the capture  
cap.release()  
cv2.destroyAllWindows()  
```

### Face and Eye Detection using HAAR Cascade Classifier

```python
# Defining the functions that will do the detections.
# load the 'haarcascade_frontalface_default.xml' file to classifiers as 'face_cascade'
face_cascade = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_frontalface_default.xml')

# load the 'haarcascade_eye.xml' file to classifiers as 'eye_cascade'
eye_cascade = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_eye.xml')


def detect(gray, frame):
    # Tuning Cascade Classifier - detectMultiScale('Input Image', "Scale Factor", "Minimum Neighbors")
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(resized_image, (x,y), (x+w, y+h), (255,0,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3)
        
        #draw bounding boxes around detected features
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
            
    return frame

# Doing some Face Recognisation with the webcam.
video = cv2.VideoCapture(0)

while True:
    check, frame = video.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow('video', canvas)
    if cv2.waitKey(1)==ord('q'):
        break
        

# When everything done, release the capture  
cap.release()  
cv2.destroyAllWindows()      
    
```

### Pedistrian Detection from the given video

```python
# import required libraries
import cv2  
import numpy as np  

# Create our body classifier
body_classifier = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_fullbody.xml')

# Initiate video capture for video file
cap = cv2.VideoCapture('Dataset/image_examples/walking.avi')

# Loop once video is successfully loaded 
while cap.isOpened():
    
    # Read first frame
    check, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Pass frame to our body classifier
    bodies = body_classifier.detectMultiScale(gray, 1.2, 3)
    
    # Extract bounding boxes for any bodies identified
    for (x, y, w, h) in bodies:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2.imshow('Pedestrians', frame)
        
    if cv2.waitKey(1)==ord('q'):
        break
        
        
# When everything done, release the capture  
cap.release()  
cv2.destroyAllWindows()           
    
```

### Car Detection from the given video

```python
# import required libraries
import cv2  
import numpy as np  
import time

# Create our body classifier
car_classifier = cv2.CascadeClassifier('Dataset/Haarcascades/haarcascade_car.xml')

# Initiate video capture for video file
cap = cv2.VideoCapture('Dataset/image_examples/cars.avi')

# Loop once video is successfully loaded 
while cap.isOpened():
    
    time.sleep(0.05)
    # Read first frame
    check, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Pass frame to our body classifier
    cars = car_classifier.detectMultiScale(gray, 1.4, 2)
    
    # Extract bounding boxes for any cars identified
    for (x, y, w, h) in cars:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        cv2.imshow('Cars', frame)
        
    if cv2.waitKey(1)==ord('q'):
        break
        
        
# When everything done, release the capture  
cap.release()  
cv2.destroyAllWindows()           
    
```
