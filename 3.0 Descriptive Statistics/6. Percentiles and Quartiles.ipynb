# <p style='text-align: center;'> Percentile </p>

### Percentile :
- A percentile is a measure at which that percentage of the total values are the same as or below that measure.


- Percentile meaning is how many of the values are less than the given percentile.


- Percentiles divide a data set into 100 equal parts. A percentile is simply a measure that tells us what percent of the total 
  frequency of a data set was at or below that measure.
  
  
- To calculate percentiles, the data must be ordered from smallest to largest.
  
  
- <b> Formula for calculating the Pth percentile's location and it's value.
    
    
                         P
                   i = ----- (n + 1)
                        100
    
    
- Where, 
    
    P : is the Pth percentile. It may or may not be part of the data.
    n : is the number of values.
    i : is the index of the value.
  
    
--------------------------------
<b> Examples :
--------------------------------
Q) We are having 15 students marks data out-of 50 marks, calculate the 50th percentile's location and it's value ?
    
    
      Marks = 34, 20, 35, 39, 38, 25, 43, 36, 31, 36, 35, 38, 15, 34, 36.
    
    
<b> Ans : -

- Number of values (n) = 15
    
    
- Pth percentile (P) = 50th percentile
    
    
- Ordering the Marks in ascending order :
    
    
     Marks = 15, 20, 25, 31, 34, 34, 35, 35, 36, 36, 36, 38, 38, 39, 43.
    
    
<b> Calculating the 50th percentile's location and it's value.
    
    
                         P
                   i = ----- (n + 1)
                        100
    
    
                        50
                   i = ----- (15 + 1)
                        100
    
    
                             
                   i = 8
    
<b> 8th index position value is 35, So 50th percentile's location = 8 and value = 35. The 50th percentile value is 35 Marks.
                        

# <p style='text-align: center;'> Quartiles </p>

### Quartiles :
- Quartiles are numbers that separate the data into quarters.


- Quartiles may or may not be part of the data.


- Q1 : The first quartile, Q1, is the same as the 25th percentile.


- Q2 : The second quartile, Q2, is the same as the 50th percentile and equals to the Median.


- Q3 : The third quartile, Q3, is the same as the 75th percentile.


---------------------
<b> Examples :
---------------------
Q) For the following 8 data-set, calculate the 25th(Q1), 50th(Q2) and 75th(Q3) Quartiles ?


           Data = 106, 129, 109, 114, 116, 121, 122, 125.
    
    
<b> Ans : -
- Number of values (n) = 8


- Ordering the data in ascending order :
    
    
           Data = 106, 109, 114, 116, 121, 122, 125, 129.
        
        
<b> Calculating the Q1 Quartile (25th percentile's) value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        25
                   i = ----- (8 + 1)
                        100
    
    
                             
                   i = 2.25
    
    
<b> 2.25 is NOT an integer. Round it down to 2 and up to 3. The data in the 2nd position is 109 and the data in the 3rd position is 114. Average 109 and 114. The 25th percentile (Q1 Quartile) is 111.5 
    
    
<b> Calculating the Q2 Quartile or 50th percentile's or Median value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        50
                   i = ----- (8 + 1)
                        100
    
    
                             
                   i = 4.5
    
    
<b> 4.5 is NOT an integer. Round it down to 4 and up to 5. The data in the 4th position is 116 and the data in the 5th position is 121. Average 116 and 121. The 50th percentile or Q1 Quartile or Median is 118.5 
    
    
<b> Calculating the Q3 Quartile (75th percentile's) value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        75
                   i = ----- (8 + 1)
                        100
    
    
                             
                   i = 6.75
    
    
<b> 6.75 is NOT an integer. Round it down to 6 and up to 7. The data in the 6th position is 122 and the data in the 7th position is 125. Average 122 and 125. The 75th percentile (Q3 Quartile) is 123.5 
    
                        

# <p style='text-align: center;'> Interquartile Range (IQR) </p>

### Interquartile Range (IQR) :
- The interquartile range is a number that indicates the spread of the middle half or the middle 50% of the data. It is the 
  difference between the third quartile (Q3) and the first quartile (Q1).
  
  
- The IQR can help to determine potential outliers.


- <b> Formula for calculating the Interquartile Range (IQR) :
    
    
                IQR = Q3 – Q1
    
    
- Where, 
    
        IQR : is Interquartile Range.
        Q3  : The third quartile, Q3, is the same as the 75th percentile. 
        Q1  : The first quartile, Q1, is the same as the 25th percentile.
    
    


-----------------------
<b> Examples :
-----------------------
Q) For the following 8 data-set, calculate the IQR ?
    
    
    Data = 106, 129, 109, 114, 116, 121, 122, 125.
    
    
    
<b> Ans : -
- Number of values (n) = 8


- Ordering the data in ascending order :
    
    
           Data = 106, 109, 114, 116, 121, 122, 125, 129.
        
        
<b> Calculating the Q1 Quartile (25th percentile's) value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        25
                   i = ----- (8 + 1)
                        100
    
    
                             
                   i = 2.25
    
    
<b> 2.25 is NOT an integer. Round it down to 2 and up to 3. The data in the 2nd position is 109 and the data in the 3rd position is 114. Average 109 and 114. The 25th percentile (Q1 Quartile) is 111.5 
    
    
<b> Calculating the Q3 Quartile (75th percentile's) value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        75
                   i = ----- (8 + 1)
                        100
    
    
                             
                   i = 6.75
    
    
<b> 6.75 is NOT an integer. Round it down to 6 and up to 7. The data in the 6th position is 122 and the data in the 7th position is 125. Average 122 and 125. The 75th percentile (Q3 Quartile) is 123.5 
    
    
<b> Calculating the Interquartile Range (IQR) :
    
    
                    IQR = Q3 – Q1
    
    
                    IQR = 123.5 - 111.5
    
    
                    IQR = 12
    
    
<b> The above example Interquartile Range (IQR) is 12.
    

# <p style='text-align: center;'> Box and Whisker Plot </p>

### Box and Whisker Plot :
- A box and whisker plot is a graphical representation of the collected data on a number line with the use of the least value, 
  greatest value and the quartiles.
  
  
- The Box and Wisker plot is mainly used to determine the outliers from the given data set.
  
  
- A box and whisker plot shows the variability of a given data set.





- From the image we can observe that there are five numbers, that is, least value, first quartile, second quartile(median), 
  third quartile and the greatest value. Therefore the box and whisker plot is also called the five number summary of a data     set.
  
  
- Quartiles divide the the data into four equal parts, the first quartile is the 25th percentile, second is the 50th percentile 
  and the third quartile is the 75th percentile of the data.

### How to Draw a Box and Whisker Plot for Data :
#### These steps will help in drawing a box and whisker plot for any given data :


- Step 1: Order the data in an ascending order and find the quartiles. 


- Step 2: Draw a number line that includes the least value and the greatest value.


- Step 3: Graph the points above the number line that represent the five points that represent, the least value, the first  
  quartile, the second quartile(median), the third quartile and the greatest value.


- Step 4: Draw the box using the value for the quartiles and draw a line through the median.


- Step 5: Draw whiskers from the box to the least and the greatest values.

# <p style='text-align: center;'> Outliers </p>
----------------------------------------------------

### Outliers :
- A potential outlier is a data point that is significantly different from the other data points. These special data points may 
  be errors or some kind of abnormality or they may be a key to understanding the data.
  
  
- The data that falls on the far left or right side of the ordered data is tested to be the outliers. Generally, the outliers 
  fall more than the specified distance from the first and third quartile. 
  
  
- Outliers are greater than Q3+(1.5 . IQR) or less than Q1-(1.5 . IQR).


- Note : The Mean is effected by the presence of the outliers.


- By using Box plot we can find the Outliers :





### How to find the Outliers for given Data-set :

- Step-1 : Sort your data from low to high.


- Step-2 : Identify the first quartile (Q1), the median, and the third quartile (Q3).


- Step-3 : Calculate your IQR = Q3 – Q1.


- Step-4 : Calculate your upper fence = Q3 + (1.5 * IQR).



- Step-5 : Calculate your lower fence = Q1 – (1.5 * IQR).


- Step-6 : Use your fences to highlight any outliers, all values that fall outside your fences.

-----------------------
<b> Example :
-----------------------
Q) For the following data-set, by using the interquartile range find outliers.


    data = 25, 37, 24, 28, 35, 22, 31, 53, 41, 64, 29
    
    
<b> Ans : -
- Number of values (n) = 11


<b> Step 1: Sort your data from low to high :
    
    
     data = 22, 24, 25, 28, 29, 31, 35, 37, 41, 53, 64
    
    
<b> Step 2: Identify the median, the first quartile (Q1), and the third quartile (Q3) :
    
    
     Calculating the Q1 Quartile (25th percentile's) value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        25
                   i = ----- (11 + 1)
                        100
    
    
                             
                   i = 3
    
    
    3rd index position value is 25, So Q1 = 25.
    
    
    Calculating the Q2 Quartile or 50th percentile's or Median value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        50
                   i = ----- (11 + 1)
                        100
    
    
                             
                   i = 6
    
    
    6th index position value is 31, So Q2 (Median) = 31.
    
    
    Calculating the Q3 Quartile (75th percentile's) value.
        
        
                         P
                   i = ----- (n + 1)
                        100
    
    
                        75
                   i = ----- (11 + 1)
                        100
    
    
                             
                   i = 9
    
    
   9th index position value is 41, So Q3 = 41.
    
    
<b> Step 3: Calculate your IQR :
    
    
                  IQR = Q3 – Q1


                  IQR = 41 - 25


                  IQR = 16
    
    
<b> Step 4: Calculate your upper fence :
- The upper fence is the boundary around the third quartile. It tells you that any values exceeding the upper fence are 
  outliers.
    
    
    Upper fence = Q3 + (1.5 * IQR)
    
    
    Upper fence = 41 + (1.5 * 16)
    
    
    Upper fence = 65
    
    
<b> Step 5: Calculate your lower fence :
- The lower fence is the boundary around the first quartile. Any values less than the lower fence are outliers.
    
    
    Lower fence = Q1 – (1.5 * IQR)
    
    
    Lower fence = 25 – (1.5 * 16)
    
    
    Lower fence = 1
    
    
<b> Step 6: Use your fences to highlight any outliers :
- Go back to your sorted dataset from Step 1 and highlight any values that are greater than the upper fence or less than your 
  lower fence. These are your outliers.
    
    
                                            Upper fence = 65
                                            Lower fence = 1
    
    

    
    
    We find one outlier, 64, in your dataset.
