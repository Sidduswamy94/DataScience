#### Z-Score :
- Z-score is also known as standard score gives us an idea of how far a data point is from the mean.


- It indicates how many standard deviations an element is from the mean. Hence, Z-Score is measured in terms of standard 
  deviation from the mean.
  
  
- For example, a standard deviation of 1 indicates the value is 1 standard deviations away from the mean. In order to use a z-
  score, we need to know the population mean (μ) and also the population standard deviation (σ). 
  
  
<b> The Formula for calculating Z-Score :
    
    
            (X – μ) 
        z = --------
               σ
    
    
- where, 
        z = Z-Score, 
        X = The value of the element, 
        μ = The population mean, and 
        σ = The population standard deviation 
    
    
### How to Calculate Z-Score :
- Usually, the population mean ((μ), the population standard deviation (σ), and the observed value (x) are provided in the 
  problem statement, and substituting the same in the above Z-score equation yields us the Z-Score value. Depending upon 
  whether the given Z-Score is positive or negative, one makes use of the respective positive Z-Table or negative Z-Table 
  available online or on the back of your statistics textbook.
    
    
-------------------------------
<b> Examples :
-------------------------------
Q) Suppose that we gathered the data from last year's final chemistry exam and found that it followed a normal distribution with a Mean of 60 and a standard deviation of 10. What proportion of students scored less than 49 on the exam ?
    
    
<b> Ans : -
- Raw score/observed value = X = 49 
    
    
- Mean score = μ = 60
    
    
- Standard deviation = σ = 10
    
    
<b> By applying the formula of z-score,  
    
    
             (X – μ) 
        z = --------
               σ
    
    
           (49 – 60) 
        z = --------
               10
    
    
                 
        z = -1.1
    
    
<b> This means that your z-score is -1.1. Since the Z-Score is Negative -1.1, we will make use of the Negative Z-Table. 
    
    
- Follow the instruction below to find the probability from the table :

        Here,  z-score = -1.1

        Firstly, map the first two digits -1.1 on the Y-axis.
        Then along the X-axis, map 0.00
        Join both axes. The intersection of the two will provide you the Z-Score value you’re looking for
    
    

    
    
- As a result, you will get the final value which is 0.1357.

- We need to convert the Z-Score into a percentage value.

        0.1357 * 100 = 13.57%

<b> So the proportion of students scored less than 49 on the exam is 0.1357 and 13.57% students scored less than 49.

Q) When measuring the heights of all students at alocal university, it was found that it was normally distributed with a Mean height of 5.5 feet, a standard deviation of 0.5 feet. What proportion of students are between 5.81 feet, and 6.3 feet tall ?


<b> Ans : -
- Minimum Raw score/observed value (X1) = 5.81 feet
    
    
- Maximum Raw score/observed value (X2) = 6.3 feet
    
    
- Mean score = μ = 5.5 feet
    
    
- Standard deviation = σ = 0.5 feet
    
    
<b> formula calculating the z-score for (X1) = 5.81 feet,  
    
    
             (X – μ) 
        z = --------
               σ
    
    
           (5.81 – 5.5) 
        z = -----------
                0.5
    
    
                 
        z = 0.62
    
    
<b> This means that your z-score for (X1) = 5.81 feet is 0.62. Since the Z-Score is Positive 0.62, we will make use of the Positive Z-Table. 
    
    
- Follow the instruction below to find the probability from the table :

        Here,  z-score = 0.62

        Firstly, map the first two digits 0.6 on the Y-axis.
        Then along the X-axis, map 0.02
        Join both axes. The intersection of the two will provide you the Z-Score value you’re looking for
    
    

    
    
- The probability for z-score = 0.62, from the table is 0.7324.
 
    
    
<b> formula calculating the z-score for (X2) = 6.3 feet, 
    
    
             (X – μ) 
        z = --------
               σ
    
    
           (6.3 – 5.5) 
        z = -----------
                0.5
    
    
                 
        z = 1.6
    
    
<b> This means that your z-score for (X2) = 6.3 feet is 1.6. Since the Z-Score is Positive 1,6, we will make use of the Positive Z-Table. 
    
    
- Follow the instruction below to find the probability from the table :

        Here,  z-score = 1.6

        Firstly, map the first two digits 1.6 on the Y-axis.
        Then along the X-axis, map 0.00
        Join both axes. The intersection of the two will provide you the Z-Score value you’re looking for
    
    

    
    
- The probability for z-score = 1.6, from the table is 0.9452.
    
    
- For getting proportion of students are between 5.81 feet, and 6.3 feet tall, we are subtracting 6.3 and 5.81 probabilities :
    
    
    0.9452 - 0.7324 = 0.2128
    
    
- As a result, you will get the final value which is 0.2128.

- We need to convert the probability into a percentage value.

        0.2128 * 100 = 21.28%

<b> So the proportion of students are between 5.81 feet, and 6.3 feet tall is 0.1357 and 21.28% students are between 5.81 feet, and 6.3 feet.
