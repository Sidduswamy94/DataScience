# <p style='text-align: center;'> Covariance </p>

### Covariance :
- Covariance defines the changes between the two variables, such that change in one variable is equal to change in another 
  variable.
  
  
- Covariance is measured in units, which are calculated by multiplying the units of the two variables.


- Covariance works on direction.


- <b> Formula for calculating Population Covariance :
    
                                 ∑ (xi – x̅)(yi - y̅)
                      Cov(x,y) = -------------------
                                         N
    
    
- Where, 
       xi = data value of x
       yi = data value of y
        x̄ = mean of x
        ȳ = mean of y
        N = number of data values.

    
- <b> Formula for calculating Sample Covariance :
    
                                 ∑ (xi – x̅)(yi - y̅)
                      Cov(x,y) = -------------------
                                         n-1
    
    
- Where, 
       xi = data value of x
       yi = data value of y
        x̄ = mean of x
        ȳ = mean of y
        n = number of data values.


### Types of Covariance :
- Covariance can have both positive and negative values. Based on this, it has two types :

        1) Positive Covariance.
        2) Negative Covariance.

### 1) Positive Covariance :
- If the covariance for any two variables is positive, that means, both the variables move in the same direction.
 
 
- If the values (greater or lesser) of one variable corresponds to the values of another variable, then they are said to be in 
  positive covariance.
  

-------------------
<b> Example :
-------------------
Q) Calculate the Covariance for the following data :
    
    
          X = 2, 8, 18, 20, 28, 30
        
          Y = 5, 12, 18, 23, 45, 50
            
            
<b> Ans : -
    
    
- <b> Number of observations, N = 6
    
    
- <b> Mean of X :
     
    
                             2 + 8 + 18 + 20 + 28 +30
           Mean of X, (X̅) = --------------------------
                                        6
    
    
        
                                       106
           Mean of X, (X̅) = --------------------------
                                        6
    
                                       
           Mean of X, (X̅) = 17.67
    
    
- <b> Mean of Y :
    
    
    
                             5 + 12 + 18 + 23 + 45 + 50
           Mean of Y, (Y̅) = ----------------------------
                                        6
    
    
        
                                       153
           Mean of Y, (Y̅) = --------------------------
                                        6
    
                                       
           Mean of Y, (Y̅) = 25.5
    
    
- <b> Covariance of X and Y :
    
    
                 ∑ (xi – x̅)(yi - y̅)
     Cov(x,y) = -------------------
                        N
    
    
                      
     Cov(x,y) = (1/6) [(2 – 17.67)(5 – 25.5) + (8 – 17.67)(12 – 25.5) + 
                (18 – 17.67)(18 – 25.5) + (20 – 17.67)(23 – 25.5) + 
                (28 – 17.67)(45 – 25.5) + (30 – 17.67)(50 – 25.5)]
    
    
                        
     Cov(x,y) = 157.83
                        
    
- Covariance of X and Y = 157.83, So it is a Positive Covariance.                   

### 2) Negative Covariance :
- If the covariance for any two variables is negative, that means, both the variables move in the opposite direction.


- It is the opposite case of positive covariance, where greater values of one variable correspond to lesser values of another 
  variable and vice-versa.
  
  
-------------------
<b> Example :
-------------------
Q) Calculate the Covariance for the following data :
    
    
          X = 1, 2, 3, 4, 5
        
          Y = 14, 16, 20, 10, 5
            
            
<b> Ans : -
    
    
- <b> Number of observations, N = 5
    
    
- <b> Mean of X :
     
    
                                1 + 2 + 3 + 4 + 5
           Mean of X, (X̅) = --------------------------
                                        5
    
    
        
                                       15
           Mean of X, (X̅) = --------------------------
                                        5
    
                                       
           Mean of X, (X̅) = 3
    
    
- <b> Mean of Y :
    
    
    
                              14 + 16 + 20 + 10 + 5
           Mean of Y, (Y̅) = ----------------------------
                                        5
    
    
        
                                       65
           Mean of Y, (Y̅) = --------------------------
                                        5
    
                                       
           Mean of Y, (Y̅) = 13
    
    
- <b> Covariance of X and Y :
    
    
                 ∑ (xi – x̅)(yi - y̅)
     Cov(x,y) = -------------------
                        N
    
    
                      
     Cov(x,y) = (1/5) [(1 – 3)(14 – 13) + (2 – 3)(16 – 13) + (3 – 3)(20 – 13) + 
                (4 – 3)(10 – 13) + (5 – 3)(5 – 13)]
    
    
                        
     Cov(x,y) = -4.8
    
- Covariance of X and Y = -4.8, So it is a Negative Covariance.

### Note : - 
- If cov(X, Y) is greater than zero, then we can say that the covariance for any two variables is positive and both the 
  variables move in the same direction.
  
  
- If cov(X, Y) is less than zero, then we can say that the covariance for any two variables is negative and both the variables 
  move in the opposite direction.
  
  
- If cov(X, Y) is zero, then we can say that there is no relation between two variables.

### Correlation :
- Correlation estimates the depth of the relationship between variables.


- Correlation is a constant value always and does not have any units.


- Correlation works on strength.


- <b> Formula for calculating Population Correlation :
    
    
                                   Cov(X,Y)
            Correlation, ρ(X,Y) = ---------
                                    σX σy
    
    
- Where,

        ρ(X,Y) = correlation between the variables X and Y
        Cov(X,Y) = covariance between the variables X and Y
        σX = standard deviation of the X variable
        σY = standard deviation of the Y variable
