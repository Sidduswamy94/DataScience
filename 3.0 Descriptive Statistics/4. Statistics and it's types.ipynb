# <p style='text-align: center;'> 4. Statistics and it's Types </p>

- Statistics simply means numerical data, and is field of math that generally deals with collection of data, tabulation, 
  and interpretation of numerical data.
  
  
- It is actually a form of mathematical analysis that uses different quantitative models to produce a set of experimental data 
  or studies of real life. 
  
  
- Statistics makes work easy and simple and provides a clear and clean picture of work you do on a regular basis.


# Types in Statistics :
<b> 1. Descriptive Statistics :
- Descriptive statistics uses data that provides a description of the population either through numerical calculation or graph 
  or table.
    
    
- Collecting, presenting and describing the data.
    
    
## a) Measure of Central Tendency :
- Measure of central tendency is also known as summary statistics that is used to represents the center point or a particular 
  value of a data set or sample set.
    
    
- In statistics, there are three common measures of central tendency as shown below,
    
        i) Mean.
       ii) Median.
      iii) Mode.

## b) Measures Of Dispersion / Spread :
- Dispersion in statistics is a way of describing how spread out a set of data is.
    
    
- When a data set has a large value, the values in the set are widely scattered; when it is small the items in the set are 
  tightly clustered.
    
    
- The spread of a data set can be described by a range of descriptive statistics including variance, standard deviation, and 
  interquartile range.

<b> Deviation from the Mean :
- Deviation from the Mean probably the most common measure. It tells you how spread out numbers are from the mean.
    
    
<b> Example 1 :
- The data-set is 5, 9, 16, 17, 18. Calculate the Deviation from the Mean of the given data-set?
    
    
<b> Step 1: Calculating the mean 
    
    
- x̅ =  (x1 + x2 + x3 + …… + xn) / n
    
    
- x̅ = (5 + 9 + 16 + 17 + 18) / 5
    
    
- x̅ = 13 
    
    
- The mean of the given data = 13
    
    
- Step 2: Calculating the difference between each data-point and mean.
    
    
                             ----------------------------------------------------
                                  Data-Point          Difference from mean
                             ----------------------------------------------------
    
                                      5                   5 - 13 = -8
    
                                      9                   9 - 13 = -4
    
                                      16                  16 - 13 = 3
    
                                      17                  17 - 13 = 4
    
                                      18                  18 - 13 = 5
    
    
- From the above example we can see that the Deviation from the mean is -8, -4, 3, 4, 5.
    
    
- If we add the Difference from mean any data set its answer is Zero only.

<b> Mean Absolute Deviation (MAD) :
- The mean absolute deviation (MAD) of a data-set is the average distance between each data point of the data-set and 
  the mean of data.
    
    
- it represents the amount of variation that occurs around the mean value in the data-set.
    
    
- It is also a measure of variation. 
    
    
- It is calculated as the average of the sum of the absolute difference between each value of the data-set and the mean. 
    
    
- <b> Formula for the Mean Absolute Deviation (MAD) :
    
    
                             MAD = (∑ |xi – mean| ) ÷ n 
    
    
<b> Example :
- The data-set is 5, 9, 16, 17, 18. Calculate the mean absolute deviation of the given data-set?
    
    
- <b> Step 1: Calculating the mean
    
    
- x̅ = (5 + 9 + 16 + 17 + 18) / 5
    
    
- x̅ = 13
    
    
- The mean of the given data = 13.
    
    
- <b> Step 2: Calculating the absolute difference between each data-point and mean. If we get difference value in -ve, we are going to considered as +ve.
    
    
                                ----------------------------------------------------
                                  Data-Point       Absolute Difference from mean 
                                ----------------------------------------------------
    
                                      5                   5 - 13 =  8
    
                                      9                   9 - 13 =  4
    
                                      16                  16 - 13 = 3
    
                                      17                  17 - 13 = 4
    
                                      18                  18 - 13 = 5
    
    
- <b> Step 3: Adding the Absolute Difference together
    
    
- (∑ |xi – mean| ) = 8 + 4 + 3 + 4 + 5
    
    
- (∑ |xi – mean| ) = 24
    
    
- <b> Step 4: Dividing the sum of absolute difference and the number of data-points.
    
    
- MAD =  (∑ |xi – mean|) ÷ n 
    
    
- MAD = 24/5
    
    
- MAD = 4.8
    
    
- Hence, we can conclude that, on average, each data-point is 5 distance away from the mean. 
    

## ii) Variance :
- It describes how much a random variable differs from its expected value. It entails computing squares of deviations.
    
    
<b> Deviation :
- Deviation is the difference between each element from the mean.
    
    
<b> Population Variance :
- The population variance is used to determine how each data point in a particular population fluctuates or is spread out.
    

- <b> Formula for calculating population variance :
     
    
                                      ∑ (xi – µ)^2
                                σ2 = -------------
                                           N
    
    
- Where, µ is the mean of data set.
    
    
- ∑ (xi – µ)^2 is the sum of squares of difference of each observation from mean.
    
    
- N is the total number of observations.
    
    
<b> Example :
- Calculate the variance of the population data: 5, 9, 16, 17, 18.
    
    
- <b> Step 1: Calculating the mean
    
    
- µ = (5 + 9 + 16 + 17 + 18) / 5
    
    
- µ = 13
    
    
- The mean of the given data = 13.
    
    
- <b> Step 2: Calculating the difference between each data-point and mean.
    
    
        ----------------------------------------------------------------------------------
            Data-Point          Difference from mean     Square of Difference from mean 
        ----------------------------------------------------------------------------------
    
                5                   5 - 13 = -8                     64
    
                9                   9 - 13 = -4                     16
    
                16                  16 - 13 = 3                      9
    
                17                  17 - 13 = 4                     16
    
                18                  18 - 13 = 5                     25
    
    
    
- <b> Step 3: Adding the Square of Difference from mean together
    
    
-  ∑ (xi – µ)^2 = 64 + 16 + 9 + 16 + 25
    
    
-  ∑ (xi – µ)^2 = 130
    
    
- <b> Step - 4: Using the formula for variance we get,
    
    
                                      ∑ (xi – µ)^2
                                σ2 = -------------
                                           N
    
    
                                          130
                                σ2 = -------------
                                           5
    
                                        
                                σ2 = 26.0
                                           

    
<b> Sample Variance :    
- Sample variance is used to find the average of the squared deviations from the mean.
    
    
- <b> Formula for calculating sample variance :
    
    
    
                                      ∑ (xi – x̅)^2
                                s^2 = -------------
                                           n-1
    

- Where, x̅ is the mean of data set.
    
    
- ∑ (xi – x̅)^2 is the sum of squares of difference of each observation from mean.
    
    
- n is the total number of observations.
    
    
<b> Example :
- Calculate the variance of the sample data: 2398, 1844, 1539, 1311.
    
    
- <b> Step 1: Calculating the mean
    
    
- x̅ = (2398 + 1844 + 1539 + 1311) / 4
    
    
- x̅ = 1773
    
    
- The mean of the given data = 1773.
    
    
- <b> Step 2: Calculating the difference between each data-point and mean.
    
    
      ----------------------------------------------------------------------------------
          Data-Point          Difference from mean     Square of Difference from mean 
      ----------------------------------------------------------------------------------
    
            2398               2398 - 1773 =  625                  3,90,625
    
            1844               1844 - 1773 =  71                   5,041
    
            1539               1539 - 1773 =  -234                 54,756
    
            1311               1311 - 1773 =  -462                 2,13,444
    
                                      
    
    
- <b> Step 3: Adding the Square of Difference from mean together
    
    
-  ∑ (xi – x̅)^2 = 3,90,625 + 5,041 + 54,756 + 2,13,444
    
    
-  ∑ (xi – x̅)^2 = 6,63,866
    
    
- <b> Step - 4: Using the formula for variance we get,
    
    
                                      ∑ (xi – x̅)^2
                               s^2 = -------------
                                          n-1
    
    
                                       6,63,866

                               s^2 = 2,21,288.67

## Coefficient of Variation :
- Coefficient of Variation is the ratio of the Standard Deviation to the mean, expressed as a percentage.


- Measurement as relative dispersion.


- The coefficient of variation is a dimensionless quantity and is usually given as a percentage. It helps to compare two data 
  sets on the basis of the degree of variation.
  
  
- <b> Formula for the Population coefficient of variation :
