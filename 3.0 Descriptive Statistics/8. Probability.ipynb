# <p style='text-align: center;'> Probability </p>

<b> Variable :
- The word 'Probability' means the chance of occurring of a particular event.


- It is generally possible to predict the future of an event quantitatively with a certain probability of being correct. 


- The probability is used in such cases where the outcome of the trial is uncertain.


- Examples : - Rolling a die, Tossing a coin, etc..
    
    
<b> Random Variable :
- It is an experiment in which all the possible outcomes of the experiment are known in advance. But the exact outcomes of any 
  specific performance are not known in advance.
    
    
- Examples : - Rolling a die, Tossing a coin, Drawing a card from a pack of 52 cards. etc..  
    
    
- <b> Formula for calculating Probability :
                 
                                 Number of interested events
                  Probability = -----------------------------
                                  Total Number of outcomes
    
    
- Always probability value lies between 0 to 1.
    
    
- Sum of all / Total probability = 1
    
    
- Minimum probability = 0
    
    
- Note : - For continous data - probability of the single value is Zero and for the probability of the interval values are some 
values are there.

- When calculating probability, there are two rules to consider when determining if two events are independent or dependent and 
  if they are mutually exclusive or not.
  
          1) Multiplication Theorem.
          2) Addition Theorem.
  

### 1) Multiplication Theorem :
Q) If A and B are two independent events, then the probability that both will occur is equal to the product of their individual probabilities.
- If A and B are two events defined on a sample space, then :


                       P(A AND B)=P(B)P(A|B) ---------------(1.1)
                           
                           
- This rule may also be written as :
  
  
                                 P(A AND B)
                       P(A|B)=  ------------
                                    P(B)
                                    
                                    
- (The probability of  A  given  B  equals the probability of  A  and  B  divided by the probability of  B .)


- If  A  and  B  are independent, then


                       P(A|B)=P(A).
                       
                       
- and Equation  1.1  becomes


                       P(A AND B)=P(A)P(B).

### 2) Addition Theorem :
Q) If A and B are two events that are not mutually exclusive, then P(A ∪B)=P(A)+P(B)- P (A∩B) ?


- If A and B are defined on a sample space, then :
  
  
                  P(A OR B)=P(A)+P(B)−P(A AND B) -------- (1.2)
                  
                  
- If A and B are mutually exclusive, then


                  P(A AND B)=0.
                  
                  
- and Equation  1.2  becomes
                
                
                  P(A OR B)=P(A)+P(B).
