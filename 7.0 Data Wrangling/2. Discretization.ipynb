## 2. Discretization :
- Discretization is the process of transforming continuous variables into discrete variables by creating a set of continuous 
  intervals that span the range of the variables values.
  
  
- Discretization is also called binning, where bin is an alternative name for interval.


- Discretization helps handle outliers by placing these values into the lower or highr intervals, together with the remaining 
  inlier values of the distribution. Thus, these outlier observations no longer differ from the rest of the values at the tails 
  of the distribution, as they are now all together in the same interval / bucket.
  
  
- In addition, by creating appropriate bins or intervals, discretization can help spread the values of a skewed variable across 
  a set of bins with equal number of observations.
  

```python
# Importing the necessary libraries.
import pandas as pd

''' Loading "stroke prediction.csv" data into a pandas dataframe, store
    in DataFrame nams as "df" and printing out its sample.'''
df = pd.read_csv("stroke prediction.csv")

# print the top 5 records with variables.
df.head()
```

```python
# Get the size of data points by printing its shape.
df.shape
```

- df DataFrame consists of 43400 rows and 12 columns.

```python
# Get the counts unique values from "stroke" variable by using value_counts().
df["stroke"].value_counts()
```

-  Counted Patients that suffer from stroke is 783 patients and those with did not suffer from stroke is 42617.

```python
# WAPP to demonstrate discretization technique.

# Create bins for the "age" attribute.
intervals = [0, 12, 19, 30, 60, 90]

# Replace the actual data values with the bins label.
categories = ["child", "teenager", "young_adult", "middle_aged", "senior_citizen"]

'''Apply discretization using intervals (bins) and the same is 
   stored in new column called “age_category”.'''

df["age_category"] = pd.cut(df["age"], bins=intervals, labels=categories)

# print the result with sample dataset.
df.head()
```

- In the above example, we have Created bins for the "age" attribute as "intervals" and Replaced the actual data values with 
  the bins label like : categories=["child", "teenager", "young_adult", "middle_aged", "senior_citizen"]. Then applied the 
  discretization using intervals (bins), labels=categories and the same is stored in new column called “age_category”.
  
  
- Here we can see that, we have created one new column as "age_category" with different categories based on the "age"  
  intervals, as foolows :
  
      age(bins)          category
      ---------          --------
      0 to 12            child
      12 to 19           teenager
      19 to 30           young_adult
      30 to 60           middle_aged
      60 to 90           senior_citizen
