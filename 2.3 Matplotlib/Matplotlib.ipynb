## Creating different types of graph :

## 1) Histogram : -
- A histogram shows continuous data in ordered rectangular columns. Usually, there are no gaps between the columns.


- The histogram displays a frequency distribution (shape) of a data set. At first glance, histograms look alike to bar graphs.   However, there is a key difference between them. Bar Chart represents categorical data and histogram represent continuous 
  data.
  
  
- Histogram uses when you want to see whether the outputs of two or more processes are different.


- Histogram uses to check whether data is skewed or normal and to check how the data is distributed.


- Histogram is a continous frequency distribution and Univariate Analysis plot.


- Important parameters in Histogram :
 
 
    1) data
    2) bins = 10 (By default bins=10)


------------------
<b> Examples :
------------------

```python
''' WAPP to plot the histogram represents count / frequency Vs Grade 
    from "df" dataframe.'''

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot a simple histogram represents count / frequency Vs Grade from "df" dataframe.
plt.hist(df["Grade"])

# Set the label for the x-axis.
plt.xlabel("Grade")

# Set the label for the y-axis.
plt.ylabel("count/frequency")

# Set title of the plot
plt.title("Histogram of Grade")

# Display all figures.
plt.show()

```

- <b> In the above example, we can read from the histogram of Grade with there counts.

## 2) Box Plot : -
- A box and whisker plot is a graphical representation of the collected data on a number line with the use of the least value, greatest value and the quartiles.


- The Box and Wisker plot is mainly used to determine the outliers from the given data set.


- In the Box Plot, the x-axis indicates the data to be plotted while the y-axis denotes the frequency distribution.


- Box Plot is a continous frequency distribution and Univariate Analysis plot.


- Box Plot used on continous data.


- Box Plot is also know as Five Number Summary.


- Quartiles divide the the data into four equal parts, the first quartile is the 25th percentile, second is the 50th percentile and the third quartile is the 75th percentile of the data.

---------------------------
<b> Examples : 
---------------------------

```python
# WAPP to plot the Box Plot on Grade from "df" dataframe.

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot the Box Plot on Grade from "df" dataframe.
plt.boxplot(df["Grade"])

# Set title of the plot
plt.title("Boxplot of Grade")

# Display all figures.
plt.show()
```

- <b> In the above example, we can read from the Boxplot of Grade, there is one outlier is there.

## 3) Scatter plot : -
- A scatter plot is a diagram where each value in the data set is represented by a dot.


- The scatter plot is an X-Y diagram that shows a relationship between two variables.


- It is used to plot data points on a vertical and a horizontal axis. The purpose is to show how much one variable affects 
  another.
  
  
- Usually, when there is a relationship between 2 variables, the first one is called independent. The second variable is called 
  dependent because its values depend on the first variable.
  
  
- Scatter plots also help you predict the behavior of one variable (dependent) based on the measure of the other variable 
  (independent).
  
  
- Scatter plot uses when having paired numerical data.


- Scatter plot uses when working with root cause analysis tools to identify the potential for problems.


- Scatter plot uses when you just want to visualize the correlation between 2 large datasets without regard to time.


- Scatter plot is considering 2 continous variables and it is a Bivariate Analysis plot.


- Scatter plot uses to check the i) Direction  ii) Strength   iii) linearity of the variables.


- Important parameters in Scatter plot :


        1) X-axis values
        2) Y-axis values
        3) color (default=blue)
        4) marker(default="o")
        5) markersize

---------------------------
<b> Examples : 
---------------------------

```python
# WAPP to plot the Scatter plot on StudyHours Vs Grade from "df" dataframe.

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot the Scatter plot on StudyHours Vs Grade from "df" dataframe.
# Set the color="red".
# Set the marker="*".
# Set the size (s)=30.
plt.scatter(df["StudyHours"], df["Grade"], color="red", marker="*", s=30)

# Set title of the plot
plt.title("StudyHours Vs Grade Scotter plot")

# Display all figures.
plt.show()
```

```python
# WAPP to plot the Scatter plot on StudyHours Vs Grade from "df" dataframe.

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot the Scatter plot on StudyHours Vs Grade from "df" dataframe.
# Set the color="blue".
# Set the marker="o".
# Set the size (s)=30.
plt.scatter(df["StudyHours"], df["Grade"], color="blue", marker="o", s=30)

# Set title of the plot
plt.title("StudyHours Vs Grade Scotter plot")

# Display all figures.
plt.show()
```

## 4) Line Plot : -
- A line chart graphically displays data that changes continuously over time.


- line Plot uses to represent the time series data, as time changes how the value changes.


- It is plotted using the pot() function.


- Each line graph consists of points that connect data to show a trend (continuous change). Line graphs have an x-axis and
  a y-axis. In the most cases, time is distributed on the horizontal axis.
    
    
- Line Plot uses when you want to show trends. For example, how house prices have increased over time.


- Line Plot uses when you want to make predictions based on a data history over time.


- Line plot uses when comparing two or more different variables, situations, and information over a given period of time.


- Note : When plotting line plot make sure that the x-axis values should be in the sequential order.


- Important parameters in Scatter plot :


    1) color (default=blue)
    2) linewidth
    3) linestyle
     

-------------------
<b> Examples :
-------------------

```python
# WAPP to plot the Line Plot  on StudyHours from "df" dataframe.

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot the Line Plot  on StudyHours from "df" dataframe.
# Set the color="green".
# Set the linewidth=3.
plt.plot(df["StudyHours"], color="blue", linewidth=3)

# Set the label for the x-axis.
plt.xlabel("Index Numbers")

# Set the label for the y-axis.
plt.ylabel("StudyHours")

# Set title of the plot
plt.title("Line Plot")

# Display all figures.
plt.show()
```

## 5) Pie Plot : -
- Whenever we want to represent the chart in percentages, then we will select pie chart. 


- When you want to create and represent the composition of something.


- Pie Chart / Plot uses to represent each categories with there percentages.


- It is very useful for displaying nominal or ordinal categories of data.


- When comparing areas of growth within a business such as profit.


- Pie charts work best for displaying data for 3 to 7 categories.


- It is a univariate analysis (Analysis of single variable).

-------------------
<b> Examples :
-------------------

```python
# WAPP to plot the Pie Plot on below data.

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# data to display on plots.
# Create the education data and store it in "education" variable.
education = ["B.Tech", "M.Tech", "B.Sc", "M,Sc"]
# Create the no_of_students data and store it in "no_of_students" variable.
no_of_students = [50, 25, 15, 10]

# Plot the Pie Plot on "no_of_students" and "education".
# Display the percent value with single decimal by using "autopct".
# Make the "B.Sc" wedge of the pie chart to stand out by using "explode".
plt.pie(no_of_students,labels=education,autopct="%1.1f%%",explode=[0,0,0.1,0])

# Display all figures.
plt.show()
```

## Stacked Bar Plot : -
- Stacked bar plots represent different groups on the highest of 1 another.


- The peak of the bar depends on the resulting height of the mixture of the results of the groups. It goes from rock bottom to the worth rather than going from zero to value.


- It is useable for decreasing data sets only not for ascending data sets.

```python
# WAPP to plot the Stacked Bar Plot on below data.

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# data to display on plots.
# Create the company data and store it in "company" variable.
company = ["GOOGL", "AMZN", "MSFT", "FB"]
# Create the company revenue data and store it in "revenue" variable.
revenue = [90, 136, 89, 27]
# Create the company profit data and store it in "profit" variable.
profit = [40, 20, 34, 12]

# Plot the Stacked Bar Plot "company" Vs "revenue".
plt.bar(company, revenue, label="Revenue")
# Plot the Stacked Bar Plot "company" Vs "profit".
plt.bar(company, profit, label="Profit")

# Function add a legend.profit
plt.legend()

# Display all figures.
plt.show()
```

```python
# WAPP to plot the seperated Stacked Bar Plot on below data.

# import numpy as np
import numpy as np

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# data to display on plots.
# Create the company data and store it in "company" variable.
company = ["GOOGL", "AMZN", "MSFT", "FB"]
# Create the company revenue data and store it in "revenue" variable.
revenue = [90, 136, 89, 27]
# Create the company profit data and store it in "profit" variable.
profit = [40, 20, 34, 12]
# Create the list of values of length of company values.
xpos = np.arange(len(company))

# Plot the Stacked Bar Plot "company" Vs "revenue".
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue")
# Plot the Stacked Bar Plot "company" Vs "profit".
plt.bar(xpos+0.2, profit, width=0.4, label="Profit")

# set the current tick location as "company" on the x-axis.
plt.xticks(xpos, company)

# Set the label for the y-axis.
plt.ylabel("Revenue(Bln)")

# Set title of the plot
plt.title("US Technology Stocks")

# Function add a legend.profit
plt.legend()

# Display all figures.
plt.show()
```

- <b> In the above example Revenue wise Amozan is high and Profit wise Google is high.

## 7) Sub Plot : -
- The subplot() function is used to draw a multiple plots in one figure.


- he subplot() function takes three arguments that describes the layout of the figure.


- The layout is organized in rows and columns, which are represented by the first and second argument.
  The third argument represents the index of the current plot.

----------------
<b> Examples :
----------------

```python
# WAPP for drawing multiple bar plots on one image by using subplot().

# import numpy as np
import numpy as np

# import pandas as pd
import pandas as pd

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Creating program for drawing multiple bar plots on one image.
ecommerce=['Snapdeal', 'Alibaba', 'Amazon', 'Flipkart']
Q1_Profit=[45, 100, 70, 40]              # Creating Q1_Profit data.
Q2_Profit=[40, 105, 65, 45]              # Creating Q2_Profit data.
Q3_Profit=[42, 120, 72, 50]              # Creating Q3_Profit data.
Q4_Profit=[34, 115, 60, 48]              # Creating Q4_Profit data.

# create a new figure.
# Creating different bar charts on one image using subplot() function.
plt.figure(figsize=(10,10))

# Creating bar chart in first plot of figure having 3 rows and 2 columns.
plt.subplot(3,2,1)
plt.bar(ecommerce, Q1_Profit)    # Creating bar chart with ecommerce Vs Q1_Profit.
plt.title("Quarter1 Profit")     # Set title of the plot.

# Creating bar chart in second plot of figure having 3 rows and 2 columns.
plt.subplot(3,2,2)
plt.bar(ecommerce, Q2_Profit)    # Creating bar chart with ecommerce Vs Q2_Profit.
plt.title("Quarter2 Profit")     # Set title of the plot.

# Creating bar chart in third plot of figure having 3 rows and 2 columns.
plt.subplot(3,2,3)
plt.bar(ecommerce, Q3_Profit)    # Creating bar chart with ecommerce Vs Q3_Profit.
plt.title("Quarter3 Profit")     # Set title of the plot.

# Creating bar chart in sixth plot of figure having 3 rows and 2 columns.
plt.subplot(3,2,6)
plt.bar(ecommerce, Q4_Profit)    # Creating bar chart with ecommerce Vs Q4_Profit. 
plt.title("Quarter4 Profit")     # Set title of the plot.

# Dispaying the charts.
plt.show()
```

- <b> In the abive example, we have plotted multiple bar plots in one image and easily we can read that as follows : 
    
    
           In Quarter1 Profit wise Alibaba is high.
           In Quarter2 Profit wise Alibaba is high.
           In Quarter3 Profit wise Alibaba is high.
           In Quarter4 Profit wise Alibaba is high.
